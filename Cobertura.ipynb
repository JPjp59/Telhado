# **Tarefa de dimencionamento de telhado**

##Dados:

image.png

```python
largura = 6
beiral_largura = 0.5

comprimento = 16
beiral_comprimento = 0.5
```

## a) A inclinação que deverá ter o cobertura

```python
inclinacao = 30
```

## b) Altura do pendural

```python
import math

inclinacao_radianos = inclinacao * (math.pi / 180)

pendural = math.tan(inclinacao_radianos)*(largura/2)

print(f"{pendural:,.2f} m")
```

## c) A área do telhado

```python
area_do_telhado = (((largura/2) + beiral_largura)/(math.cos(inclinacao_radianos))) * (comprimento + (beiral_comprimento*2)) * 2
                  #

print(f"{area_do_telhado:,.2f} m²")
```

## d) O número de telhas

image.png

```python
rendimento_de_telhas_m2 = 17

n_de_telhas = area_do_telhado*rendimento_de_telhas_m2

n_de_telhas_arredondado = math.ceil(n_de_telhas)
print(f"{n_de_telhas_arredondado:,.0f} telhas")
```

```python
r = n_de_telhas_arredondado*1.05

r = math.ceil(r)

r
```

## e) O peso aproximado do telhado

image.png

```python
peso_do_telhado = (125+60)*area_do_telhado

print(f"{peso_do_telhado:,.2f} kg")
```

## f) O número de tesouras e a distância entre tesouras

image.png

```python
distancia_entre_tesouras = 3.5

n_de_tesouras = (comprimento/distancia_entre_tesouras)+1

n_de_tesouras_arredondado = math.ceil(n_de_tesouras)

n_de_tesouras_arredondado

# Recalculando

novo_espacamento_entre_tesouras = comprimento/(n_de_tesouras_arredondado -1)

print(f"O número de tesoura é de: {n_de_tesouras_arredondado}, e o espaçamento entre elas é de: {novo_espacamento_entre_tesouras:.2f}m.")
```

II. Cumeeira

```python
comprimento_cumeeira = comprimento+(beiral_comprimento*2)

# Determina a secção:
if largura_com_beiral > 5:
    seccao_cumeeira = "7,5 x 23 cm"
else:
    seccao_cumeeira = "7,5 x 15 cm"

print(f"comprimento da cumeeira: {comprimento_cumeeira}")
print(f"Secção da cumeeira: {seccao_cumeeira}")
```

III. Caibros

```python
espacamento_entre_caibros = 0.5 #peguei da tabela da apostila.

n_de_caibros_p_um_lado = ((comprimento + (beiral_comprimento*2))/espacamento_entre_caibros)+1

n_de_caibros = n_de_caibros_p_um_lado*2

# Determina a secção:
if largura_com_beiral > 5:
    seccao_caibros = "7,5 x 7,5 cm"
else:
    seccao_caibros = "3,8 x 7,5 cm"

# Determina o comprimento:
comprimento_caibros = math.sqrt(((largura/2)**2)+((pendural)**2))

#Calculado hipotenusa apenas para o beiral

hipotenusa = 0.5/math.cos(inclinacao_radianos)


print(f"O número de caibros é de: {n_de_caibros:,.0f}")
print(f"Secção dos caibros: {seccao_caibros}")
print(f"comprimento dos caibros: {comprimento_caibros+hipotenusa:.2f} m")
```

IV. Ripas

```python
n_de_ripas = ((comprimento_caibros+hipotenusa)/0.32)+1

n_de_ripas = math.ceil(n_de_ripas)

n_de_ripas = n_de_ripas*2

print(f"O número de ripas é de {n_de_ripas:.0f} com uma seção transversal de: 1,5 x 5cm")
```
